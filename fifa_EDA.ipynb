# Imputation of missing values

```python
fifa['player_traits'].fillna("no data", inplace=True)
```

```python
fifa['player_traits'].isnull().sum()
```

```python
#fifa['physic'].fillna(method="ffill" , inplace=True)
```

```python
fifa['physic'].fillna(fifa.physic.median() , inplace=True)
fifa['defending'].fillna(fifa.defending.median(), inplace=True)
```

```python

fifa['dribbling'].fillna(method="ffill" , inplace=True)
fifa['passing'].fillna(method="ffill" , inplace=True)
fifa['shooting'].fillna(method="ffill" , inplace=True)
fifa['pace'].fillna(method="ffill" , inplace=True)
fifa['release_clause_eur'].fillna(method="ffill" , inplace=True)
fifa['club_joined'].fillna(method="ffill" , inplace=True)
fifa['league_level'].fillna(method="ffill" , inplace=True)
fifa['value_eur'].fillna(method="ffill" , inplace=True)
fifa['club_jersey_number'].fillna(method="ffill" , inplace=True)
fifa['club_flag_url'].fillna(method="ffill" , inplace=True)
fifa['club_logo_url'].fillna(method="ffill" , inplace=True)
fifa['club_team_id'].fillna(method="ffill" , inplace=True)
fifa['club_name'].fillna(method="ffill" , inplace=True)
fifa['league_name'].fillna(method="ffill" , inplace=True)
fifa['club_position'].fillna(method="ffill" , inplace=True)
fifa['wage_eur'].fillna(method="ffill" , inplace=True)
fifa['club_contract_valid_until'].fillna(method="ffill" , inplace=True)
```
